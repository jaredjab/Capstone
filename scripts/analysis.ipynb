```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
acreage = pd.read_csv('../data/final/acreage.csv')
biodiversity = pd.read_csv('../data/final/biodiversity.csv')
climate = pd.read_csv('../data/final/climate.csv')
fees = pd.read_csv('../data/final/fees.csv')
general = pd.read_csv('../data/final/general.csv')
visits = pd.read_csv('../data/final/visits.csv')
```

```python
biodiversity
```

```python
agg_bio = biodiversity.groupby(['park_name'])['common_names'].count().to_frame().reset_index().rename(columns={'common_names':'total_species'})
```

```python
agg_bio.head(1)
```

```python
agg_bio_cats = biodiversity.groupby(['park_name', 'category'])['common_names'].count().unstack().fillna(0)
```

```python
agg_bio_cats
```

```python
agg_bio_cats.at['Acadia', 'Amphibian']
```

```python
categories = biodiversity.category.unique()
categories
```

```python
for cat in categories:
    agg_bio[cat] = 0
```

```python
agg_bio
```

```python
for index, fields in agg_bio.iterrows():
    for cat in categories:
        agg_bio.at[index, cat] = agg_bio_cats.at[fields.park_name, cat] if (fields.park_name in agg_bio_cats.index) else np.nan
```

```python
agg_bio = agg_bio[['park_name', 'Mammal', 'Bird', 'Reptile', 'Amphibian', 'Fish', 'Vascular Plant']]
```

```python
agg_bio['total_species'] = agg_bio.drop(columns='park_name').sum(axis=1)
```

```python
agg_bio
```

```python
agg_bio.to_csv('../data/final/bio_by_park.csv', index=False)
```

```python
visits.loc[visits.Year == 2023].groupby('park_name').sum().reset_index().drop(columns=['Year', 'Month'])
```

```python
for col in visits.columns:
    if col not in ['Year', 'Month', 'park_name', 'NonRecreationOvernightStays', 'ConcessionerCamping']:
        visits[col] = pd.to_numeric(visits[col].str.replace(',', ''), errors='coerce')
```

```python
visits
```

```python
agg_visits = visits.loc[visits.Year == 2023].groupby('park_name').sum().reset_index().drop(columns=['Year', 'Month'])
```

```python
agg_visits.info()
```

```python
agg_bio.info()
```

```python
visits_and_bio_2023 = agg_visits.merge(agg_bio
                ,how='left'
                ,left_on='park_name'
                ,right_on='park_name')
```

```python
v_b_corr = visits_and_bio_2023.corr(numeric_only=True)
```

```python
v_b_corr = v_b_corr.reset_index().rename(columns={'index':'Category'})
```

```python
v_b_corr.to_csv('../data/final/visits_bio_corr.csv')
```
