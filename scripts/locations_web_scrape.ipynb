```python
import requests
from bs4 import BeautifulSoup as BS
import pandas as pd
import numpy as np
import re
```

```python
response = requests.get('https://www.latlong.net/category/national-parks-236-42.html')
response.status_code
```

```python
soup = BS(response.text)
```

```python
loc_table = soup.find_all('table')
```

```python
locations = pd.read_html(str(loc_table[0]))[0]
```

```python
locations
```

```python
locations['park_name'] = locations['Place Name'].str.replace(r'National.*$', '', regex=True).str.strip()
```

```python
locations = locations.rename(columns={'Latitude':'lat', 'Longitude':'long'})
```

```python
locations = locations.drop(columns='Place Name')
```

```python
locations = locations.sort_values(by='park_name')
locations = locations.drop_duplicates('park_name').reset_index(drop=True)
locations
```

```python
locations.park_name.to_list() # Missing Parks
```

```python
np_all = pd.read_csv('../data/np_all.csv')
```

```python
parks = np_all.park_name.to_list()
```

```python
for index, fields in locations.iterrows():
    if fields.park_name in parks:
        pass
    else:
        locations = locations.drop(index)
locations = locations.reset_index(drop=True)
```
