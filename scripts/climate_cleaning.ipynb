```python
import pandas as pd
import os 
```

```python
acadia = pd.read_csv('../data/climate/acadia.csv')
```

```python
acadia
```

```python
acadia['park_name'] = 'Acadia'
```

```python
acadia
```

```python
files = os.listdir('C:/Users/Jab/Documents/NSS/Capstone/data/climate')
climate = acadia

for file in files[1:]:
    df = pd.read_csv(f'../data/climate/{file}')
    df['park_name'] = file[:-4]
    frames = [climate, df]
    climate = pd.concat(frames)

climate = climate.rename(columns={'STATION':'station'
                                 ,'DATE':'month'
                                 ,'MLY-PRCP-NORMAL':'precipitation_in'
                                 ,'MLY-SNOW-NORMAL':'snow_in'
                                 ,'MLY-TAVG-NORMAL':'avg_temp_f'
                                 ,'MLY-TMAX-NORMAL':'max_temp_f'
                                 ,'MLY-TMIN-NORMAL':'min_temp_f'})
```
