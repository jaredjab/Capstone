```python
import pandas as pd
import re
```

```python
acreage = pd.read_csv('../data/national-park-service/nps_acreage_2023.csv', encoding='latin-1')
```

```python
np_acres = acreage.loc[(acreage['Area Name'].str.endswith('NP'))
           |(acreage['Area Name'].str.endswith('NP & PRES'))
           |(acreage['Area Name'].str.endswith('BLACK CANYON OF GUNNISON'))
           |(acreage['Area Name'].str.endswith('NP OF AMERICAN SAMOA'))]
```

```python
np_acres = np_acres.drop(columns=['Region'
                               ,'Unnamed: 11'
                               ,'Unnamed: 12'
                               ,'Unnamed: 13'])
```

```python
np_acres['park_name'] = np_acres['Area Name'].str.title()
```

```python
for index, fields in np_acres.iterrows():
    if fields.park_name.endswith('Samoa'):
        np_acres.loc[index, 'park_name'] = 'American Samoa'
    elif fields.park_name.startswith('T Roosevelt'):
        np_acres.loc[index, 'park_name'] = 'Theodore Roosevelt'
    elif fields.park_name.startswith('Black Canyon'):
        np_acres.loc[index, 'park_name'] = 'Black Canyon of the Gunnison'
    elif fields.park_name.endswith(' Np'):
        np_acres.loc[index, 'park_name'] = fields.park_name[:-3]
    elif fields.park_name.endswith(' Np & Pres'):
        np_acres.loc[index, 'park_name'] = fields.park_name[:-10]
    
np_acres = np_acres.sort_values(by='park_name').reset_index(drop=True)
```

```python
np_acres = np_acres.drop(columns='Area Name')
```
