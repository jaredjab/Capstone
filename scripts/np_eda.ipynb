```python
biodiversity = pd.read_csv('../data/biodiversity.csv')
```

```python
biodiversity
```

```python
category_counts = biodiversity.groupby(['park_name', 'category']).count()
category_counts
```

```python
park_counts = biodiversity.park_name.value_counts().to_frame().sort_values(by='park_name').reset_index()
park_counts.head()
```

```python
park_counts = park_counts.rename(columns={'count':'num_species'})
park_counts.head()
```

```python
np_stats = np.merge(park_counts
                   ,how='inner'
                   ,left_on='park_name'
                   ,right_on='park_name')
np_stats.info()
```

```python
np_stats.head()
```

```python
def acresToSqMiles(acres):
    sq_miles = round(acres / 640.0, 2)
    return sq_miles
```

```python
np_stats['area_sq_miles'] = np_stats.area_acres.apply(acresToSqMiles)
```

```python
np_stats
```

```python
np_stats = np_stats.drop(columns=['area_acres', 'area_km2'])
np_stats.info()
```

```python
np_stats
```

```python
np_stats['species_per_sq_mile'] = np_stats.apply(lambda x: round(x.num_species / x.area_sq_miles, 2), axis=1)
```

```python
np_stats['rec_hours_per_sq_mile'] = np_stats.apply(lambda x: round(x.recreational_hours / x.area_sq_miles, 2), axis=1)
```

```python
np_stats['non_rec_hours_per_sq_mile'] = np_stats.apply(lambda x: round(x.non_recreational_hours / x.area_sq_miles, 2), axis=1)
```

```python
np_stats['rec_hours_per_visit'] = np_stats.apply(lambda x: round(x.recreational_hours / x.recreational_visits, 2), axis=1)
```

```python
def nonRecHrPerVisit(hours, visits):
    try:
        result = round(hours / visits, 2)
        return result
    except:
        return 0.0
```

```python
np_stats['non_rec_hours_per_visit'] = np_stats.apply(lambda x: nonRecHrPerVisit(x.non_recreational_hours, x.non_recreational_visits), axis=1)
```

```python
np_stats
```

```python
np_stats.to_csv('../data/np_all.csv', index=False)
```
