```python
def convertNumeric(df): # Changes all columns to numeric except for the first one
    for column in df.columns[1:]:
        df[column] = pd.to_numeric(df[column].str.replace(',', ''))
    return df
```

```python
convertNumeric(visitation)
```

```python
visitation.describe()
```

```python
parks_info = pd.read_csv('../data/wikipedia/parks_info.csv')
```

```python
parks_info
```

```python
# parks_visitation_22 = visitation.loc[visitation.park_name.str.endswith(' NP')]
parks_visitation_22 = visitation
```

```python
def nameClean(name):
    if name.endswith(' NP'):
        return name[:-3]
    elif name.endswith(' NP & PRES'):
        return name[:-10]
    else:
        return name
```

```python
parks_visitation_22.park_name = parks_visitation_22.park_name.apply(nameClean)
parks_visitation_22
```

```python
parks_visitation_22.info()
```

```python
parks_info.info()
```

```python
all_parks_info = (
parks_info.merge(parks_visitation_22
                ,how='left'
                ,left_on='park_name'
                ,right_on='park_name')
)
all_parks_info.info()
```

```python
all_parks_info
```

```python
all_parks_info.to_csv('../data/np_data.csv', index=False)
```
